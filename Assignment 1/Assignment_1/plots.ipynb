```python
import matplotlib.pyplot as plt
import pandas as pd

algorithms = ['Iterative Deepening Search','Bidirectional Search','A* Search', 'Bidirectional Heuristic Search']
memory_usage = [3.19921875,2.1796875,2.62109375,2.24609375]
time_taken = [274.06890869140625,35.72912931442261,42.56570792198181,46.15577006340027]
cost = [19803790.626988005,14482908.243998483,13540779.809997834,13555830.6589979]

data = {
'Algorithm': algorithms,
    'Memory Usage (MiB)': memory_usage,
    'Time Taken (s)': time_taken,
    'Cost': cost}
df = pd.DataFrame(data)
df['Algorithm_num'] = range(len(df))

plt.figure(figsize=(10, 6))
plt.scatter(df['Algorithm_num'], df['Memory Usage (MiB)'], color='skyblue', s=100)
plt.title('Memory Usage vs Algorithm')
plt.xlabel('Algorithm')
plt.ylabel('Memory Usage (MiB)')
plt.xticks(df['Algorithm_num'], df['Algorithm'], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(df['Algorithm_num'], df['Time Taken (s)'], color='salmon', s=100)
plt.title('Time Taken vs Algorithm')
plt.xlabel('Algorithm')
plt.ylabel('Time Taken (s)')
plt.xticks(df['Algorithm_num'], df['Algorithm'], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(df['Algorithm_num'], df['Cost'], color='lightgreen', s=100)
plt.title('Cost vs Algorithm')
plt.xlabel('Algorithm')
plt.ylabel('Cost')
plt.xticks(df['Algorithm_num'], df['Algorithm'], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
