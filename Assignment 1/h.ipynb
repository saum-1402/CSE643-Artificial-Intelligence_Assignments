```python
import numpy as np

def dfs_with_depth_limit(adj_matrix, start, goal, depth_limit, current_depth, path, visited):
    if current_depth > depth_limit:  # Depth limit exceeded
        return False
    
    visited[start] = True  # Mark the node as visited
    path.append(start)  # Add node to path

    if start == goal:  # Goal reached
        return True
    
    # Loop through neighbors
    for neighbor in range(len(adj_matrix[start])):
        if adj_matrix[start][neighbor] > 0 and not visited[neighbor]:  # Check if there's an edge and the neighbor is unvisited
            if dfs_with_depth_limit(adj_matrix, neighbor, goal, depth_limit, current_depth + 1, path, visited):
                return True  # Path found

    # Backtrack
    path.pop()  # Remove the node from the path if it doesn't lead to the goal
    visited[start] = False  # Mark the node as unvisited for the next search
    
    return False

def iterative_deepening_search(adj_matrix, start, goal):
    num_nodes = len(adj_matrix)  # Number of nodes in the graph
    for depth_limit in range(num_nodes):  # Increment depth limit step by step
        visited = [False] * num_nodes  # Reset visited set for each depth
        path = []  # Reset path for each depth
        if dfs_with_depth_limit(adj_matrix, start, goal, depth_limit, 0, path, visited):
            return path  # Path found within depth limit
    return None  # No path found

# Example usage:

# Create an adjacency matrix (numpy array)
# adj_matrix = np.array([[0, 1, 1, 0, 0, 0],  # Node a (0) is connected to b (1) and c (2)
#                        [0, 0, 0, 1, 0, 0],  # Node b (1) is connected to d (3)
#                        [0, 0, 0, 0, 1, 0],  # Node c (2) is connected to e (4)
#                        [0, 0, 0, 0, 0, 1],  # Node d (3) is connected to f (5)
#                        [0, 0, 0, 1, 0, 0],  # Node e (4) is connected to d (3)
#                        [0, 0, 0, 0, 0, 0]]) # Node f (5) has no outgoing edges

adj_matrix = data

# Find the path from node 0 (a) to node 5 (f)
start_node = 12
goal_node = 67
path = iterative_deepening_search(adj_matrix, start_node, goal_node)

if path:
    print(f"Path found: {path}")
else:
    print("No path found")
```

```python
import networkx as nx
import numpy as np
import heapq

def heuristic(node, goal):
    # Example heuristic function (e.g., zero heuristic for unweighted graphs)
    return 0

def bidirectional_heuristic_search(adj_matrix, start, goal):
    # Create a NetworkX graph from the adjacency matrix
    G = nx.from_numpy_array(adj_matrix, create_using=nx.DiGraph)
    
    # Initialize search structures
    open_start = []
    open_target = []
    came_from_start = {}
    came_from_target = {}
    cost_from_start = {start: 0}
    cost_from_target = {goal: 0}
    
    heapq.heappush(open_start, (0 + heuristic(start, goal), start))
    heapq.heappush(open_target, (0 + heuristic(goal, start), goal))
    
    came_from_start[start] = None
    came_from_target[goal] = None
    
    while open_start and open_target:
        current_cost, current_node = heapq.heappop(open_start)
        if current_node in came_from_target:
            return reconstruct_path(came_from_start, came_from_target, current_node)
        
        for neighbor in G.neighbors(current_node):
            new_cost = cost_from_start[current_node] + G[current_node][neighbor].get('weight', 1)
            if neighbor not in cost_from_start or new_cost < cost_from_start[neighbor]:
                cost_from_start[neighbor] = new_cost
                priority = new_cost + heuristic(neighbor, goal)
                heapq.heappush(open_start, (priority, neighbor))
                came_from_start[neighbor] = current_node
        
        current_cost, current_node = heapq.heappop(open_target)
        if current_node in came_from_start:
            return reconstruct_path(came_from_start, came_from_target, current_node)
        
        for neighbor in G.neighbors(current_node):
            new_cost = cost_from_target[current_node] + G[current_node][neighbor].get('weight', 1)
            if neighbor not in cost_from_target or new_cost < cost_from_target[neighbor]:
                cost_from_target[neighbor] = new_cost
                priority = new_cost + heuristic(neighbor, start)
                heapq.heappush(open_target, (priority, neighbor))
                came_from_target[neighbor] = current_node
    
    return None

def reconstruct_path(came_from_start, came_from_target, meeting_node):
    # Reconstruct the path from start to the meeting point
    path_start_to_meet = []
    node = meeting_node
    while node is not None:
        path_start_to_meet.append(node)
        node = came_from_start[node]
    path_start_to_meet.reverse()
    
    # Reconstruct the path from meeting point to target
    path_meet_to_target = []
    node = came_from_target[meeting_node]
    while node is not None:
        path_meet_to_target.append(node)
        node = came_from_target[node]
    
    # Combine paths
    return path_start_to_meet + path_meet_to_target

# Example usage:
adj_matrix = data
start_node = 12
goal_node = 67
path = bidirectional_heuristic_search(adj_matrix, start_node, goal_node)

if path:
    print(f"Path found: {path}")
else:
    print("No path found")
```
